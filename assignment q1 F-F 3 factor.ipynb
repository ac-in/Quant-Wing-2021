```python
import pandas_datareader.data as reader
import pandas as pd
import datetime as dt
import statsmodels.api as sm
import fix_yahoo_finance
#import yfinance as yf
```

```python
end = dt.date(2020, 6, 30)
start = dt.date(end.year - 5, end.month, end.day)
funds = ['FDGRX']
```

```python
fundsret = reader.get_data_yahoo(funds, start, end)['Adj Close'].pct_change()
```

```python
fundsret_mtl = fundsret.resample('M').agg(lambda x: (x+1).prod() - 1)
```

```python
fundsret_mtl = fundsret_mtl[1:]
```

```python
fundsret_mtl
```

```python
factors = reader.DataReader('F-F_Research_Data_Factors', 'famafrench', start, end)[0]
```

```python
factors = factors[1:] #starting with july 
```

```python
#we are only interested in the 0:th one from the dictionary obtained if not used [0]
```

```python
fundsret_mtl.index = factors.index
```

```python
#merginf the two dataframes
```

```python
merge = pd.merge(fundsret_mtl, factors, on = 'Date')
```

```python
merge
```

```python
#last three factors are in perc so we divide by 100
```

```python
merge[['Mkt-RF', 'SMB', 'HML', 'RF']] = merge[['Mkt-RF', 'SMB', 'HML', 'RF']]/100
```

```python
merge
```

```python
merge['FDGRX-RF'] = merge.FDGRX - merge.RF
```

```python
merge
```

```python
y = merge['FDGRX-RF']
X = merge[['Mkt-RF', 'SMB', 'HML']]

X_sm = sm.add_constant(X)
```

```python
model = sm.OLS(y, X_sm)
result = model.fit()
result.summary()
```
